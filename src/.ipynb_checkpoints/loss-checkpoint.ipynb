```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.FloatTensor(np.linspace(-1, 1, 100))
weight, margin = 0.5, 1

pos = weight * (x ** 2)
neg = torch.clamp(margin - x, min=0) ** 2

plt.plot(x.numpy(), pos.numpy(), label='pos')
plt.plot(x.numpy(), neg.numpy(), label='neg')
plt.legend();
```

```python
x = torch.FloatTensor(np.linspace(-1, 1, 100))
weight, margin = 0.25, 0.5

pos = weight * ((1 - x) ** 2)
neg = (x ** 2) * torch.gt(x, margin).float()

plt.plot(x.numpy(), pos.numpy(), label='pos')
plt.plot(x.numpy(), neg.numpy(), label='neg')
plt.legend();
```

```python
x = torch.FloatTensor(np.linspace(-1, 1, 100))
weight, margin = 0.25, 0.5

pos = weight * ((1 - x) ** 2)
neg = (1 - (x ** 2)) * torch.gt(x, margin).float()

plt.plot(x.numpy(), pos.numpy(), label='pos')
plt.plot(x.numpy(), neg.numpy(), label='neg')
plt.legend();
```
