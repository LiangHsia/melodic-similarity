```python
import torch
import numpy as np

%matplotlib notebook
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
def hardest(dist):
    idx = np.argmax(dist)
    return idx if dist[idx] > 0 else None

def random_hard_negative(dists):
    hard_ones = np.where(dists > 0)[0]
    idx = None
    if len(hard_ones) > 0:
        idx = np.random.choice(hard_ones)
    return idx


def semihard_negative(dists, margin=1):
    semi_hard_ones = np.where((dists < margin) & (dists > 0))[0]
    idx = None
    if len(semi_hard_ones) > 0:
        idx = np.random.choice(semi_hard_ones)
    return idx
```

```python
x = torch.FloatTensor(np.linspace(-1, 1, 100))
y = torch.FloatTensor(np.linspace(1, -1, 100))
margin = 1

dist = x - y - margin

print(hardest(dist))
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(10, 4))
axes = axes.flatten()

for margin in np.arange(0, 1.25, 0.25):
    dist = x - y + margin
    axes[0].plot(range(x.shape[0]), dist.numpy())
    indexes = [random_hard_negative(dist) for i in range(10000)]
    print(margin, np.mean([dist[idx] for idx in indexes]))
    sns.distplot([dist[idx] for idx in indexes], hist=False, ax=axes[1])
    sns.distplot(indexes, hist=False, ax=axes[2], label=str(margin))
plt.legend();
```
