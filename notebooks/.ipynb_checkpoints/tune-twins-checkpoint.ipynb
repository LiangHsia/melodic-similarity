### Step 0: Setup paths

This notebook briefly describes the steps to run a Siamese LSTM network on the Meertens Tune Collections. First, we append the path of all scripts to python load path:

```python
import sys

sys.path.append('../src/')
```

### Step 1: Load the data

Next, we load the data, and make a split for training and development data:

```python
import json

from sklearn.model_selection import train_test_split

with open('../data/mtcann.json') as f:
    data = json.load(f)

train_data, dev_data = train_test_split(data, test_size=0.1)
```

### Step 2: Setup feature encoders

Features need to be encoded with the `Encoder` class. For each feature, initialize an instance of the class and provide it with the name of the feature.

```python
from encoders import Encoder

scale_degree_encoder = Encoder('scaledegrees')
metric_contour_encoder = Encoder('metriccontour')
```

### Step 3: Setup Data Loaders and Pair Selectors

To set up the datasets, we use the `DupletDataset`, one for the training and one for the development set. This class needs a specific example selector, which provides the dataset with pairs for training or testing. Here, we use the simple `PairSelector` class.

```python
from samplers import PairSelector
# from datasets import DupletDataset
%run ../src/datasets.py

train_selector = PairSelector(pos_neg_ratio=3)
dev_selector = PairSelector(pos_neg_ratio=0) # use all pairs for validation
train = DupletDataset(
    train_data, scale_degree_encoder, metric_contour_encoder, max_seq_length=20, 
    selector=train_selector, padding='left', batch_size=20).fit()
dev = DupletDataset(
    dev_data, scale_degree_encoder, metric_contour_encoder, max_seq_length=20,
    selector=dev_selector, padding='left', batch_size=20).fit()
print(train_selector)
print(dev_selector)
```

### Step 4: Initialize the Network

Next, we setup all models. Note that to train on the GPU, you need to set `device = "cuda"`; use `device = "cpu"` to train on the CPU. 

```python
import torch
# from models import LSTM, SiameseLSTM, ContrastiveLoss
%run ../src/models.py

device = "cpu" # change to cpu if working on a machine without GPU
network = LSTM(8, 64, scale_degree_encoder, metric_contour_encoder, n_layers=2)
model = SiameseLSTM(network, ContrastiveLoss()).to(device)
print(model)
for encoder in model.network.encoders:
    encoder.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, 'min', verbose=True, patience=5, threshold=1e-4)
```

### Step 5: Train!

Finally, train the model! 

```python
# from trainer import fit_model
%run ../src/trainer.py
from metrics import SiameseMetrics

%matplotlib inline

fit_model(train, dev, model, optimizer, scheduler, 50, device, SiameseMetrics())
```

### Step 6: Visualize Embeddings

As a final step, we can visualize the learned representations of melodies in a two-dimensional space. There are many techniques available to do this. Here, we use a t-SNE.

```python
from utils import extract_embeddings, plot_embeddings

train_embs, train_y, train_ids = extract_embeddings(train, model, device=device)
dev_embs, dev_y, dev_ids = extract_embeddings(dev, model, device=device)
```

For the t-SNE visualization, we can use something like the following:

```python
from sklearn.manifold import TSNE
import altair as alt
import numpy as np

alt.renderers.enable('notebook')

embs = np.vstack((train_embs, dev_embs))
y = np.hstack((train_y, dev_y))
ids = np.hstack((train_ids, dev_ids))
train_or_test = ['train'] * len(train) + ['validate'] * len(dev)

plot_embeddings(embs, TSNE(perplexity=10, method='exact'), labels=y, ids=ids, train_or_test=train_or_test)
```

```python
from sklearn.metrics import pairwise_distances, average_precision_score

MAP = []
dists = 1 - pairwise_distances(dev_embs, train_embs, metric='cosine')
for i, dist in enumerate(dists):
    target = dev_y[i]
    targets = (train_y == target).astype(int)
    MAP.append(average_precision_score(targets, dist))
print(np.mean(MAP))
```

### Step 7: Different pair selectors

As said, there are different ways to select pairs for training. Previously we've used the `PairSelector` class which randomly samples positive and negative pairs. Another selector is the `AlignmentBasedPairSelector` which ranks pairs on the basis of alignment scores (TODO: which ones, which system?). Below, we setup an experiment with this class, which should provide harder examples to the network. 

```python
from samplers import AlignmentBasedPairSelector
from datasets import DupletDataset

train_selector = AlignmentBasedPairSelector(pos_neg_ratio=3, alignment_path='../data/mtcann_alignmentscores.csv')
dev_selector = PairSelector(pos_neg_ratio=0) # use all pairs for validation
train = DupletDataset(
    train_data, scale_degree_encoder, metric_contour_encoder, selector=train_selector, batch_size=20).fit()
dev = DupletDataset(
    dev_data, scale_degree_encoder, metric_contour_encoder, selector=dev_selector, batch_size=20).fit()
print(train_selector)
print(dev_selector)
```

Next, we setup a new network:

```python
device = "cuda" # change to cpu if working on a machine without GPU
network = LSTM(8, 64, scale_degree_encoder, metric_contour_encoder, n_layers=2)
model = SiameseLSTM(network).to(device)
print(model)
for encoder in model.network.encoders:
    encoder.to(device)
loss_fn = ContrastiveLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, 'min', verbose=True, patience=5, threshold=1e-4)
```

And train the model:

```python
fit_model(train, dev, model, loss_fn, optimizer, scheduler, 50, device, SiameseMetrics())
```

Somewhat surprisingly, overfitting is more of an issue here. Quite possibly, this is the result of the network overestimating the importance of relatively minor differences between examples. Since it has only been given examples that are very much alike, it learns to focus on characteristics too specific to transfer to other, more easier examples. The plot below illustrates this, where separation of training classes appears to be much lower:

```python
train_embs, train_y, train_ids = extract_embeddings(train, model, device=device)
dev_embs, dev_y, dev_ids = extract_embeddings(dev, model, device=device)

embs = np.vstack((train_embs, dev_embs))
y = np.hstack((train_y, dev_y))
ids = np.hstack((train_ids, dev_ids))
train_or_test = ['train'] * len(train) + ['validate'] * len(dev)

plot_embeddings(embs, TSNE(perplexity=10, method='exact'), labels=y, ids=ids, train_or_test=train_or_test)
```

```python
MAP, MAP_at1 = [], []
dists = 1 - pairwise_distances(dev_embs, train_embs, metric='cosine')
for i, dist in enumerate(dists):
    target = dev_y[i]
    targets = (train_y == target).astype(int)
    MAP.append(average_precision_score(targets, dist))
    MAP_at1.append(ranking_precision_score(targets, dist, k=10))
print(np.mean(MAP), np.mean(MAP_at1))
```

Thus, it might be interesting to experiment with a mixed pair selector, which provides both hard and random pairs for training. The `AlignmentBasedPairSelector` allows you the specify the ratio of hard examples. Let's try and see if this improves the overfitting issue:
