```python
from ipywidgets import interact

x = torch.FloatTensor(np.linspace(-1, 1, 100))
weight, margin = 0.25, 1
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
pos = weight * ((1-x) ** 2)
neg = torch.clamp(margin - (1-x), min=0) ** 2
line_pos, = ax.plot(x.numpy(), pos.numpy())
line_neg, = ax.plot(x.numpy(), neg.numpy())

def update(margin=1.0):
    pos = weight * ((1-x) ** 2)
    neg = torch.clamp(margin - (1-x), min=0) ** 2
    line_pos.set_ydata(pos.numpy())
    line_neg.set_ydata(neg.numpy())
    fig.canvas.draw()

interact(update);
```

```python
x = torch.FloatTensor(np.linspace(-1, 1, 100))
weight, margin = 0.25, 0.4
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
pos = weight * ((1 - x) ** 2)
neg = (x ** 2) * torch.gt(x, margin).float()
line_pos, = ax.plot(x.numpy(), pos.numpy())
line_neg, = ax.plot(x.numpy(), neg.numpy())

def update(margin=0.25):
    pos = weight * ((1 - x) ** 2)
    neg = (x * torch.gt(x, margin).float()) ** 2
    line_pos.set_ydata(pos.numpy())
    line_neg.set_ydata(neg.numpy())
    fig.canvas.draw()

interact(update);
```

```python
from ipywidgets import interact

x = torch.FloatTensor(np.linspace(-1, 1, 100))
y = torch.FloatTensor(np.linspace(1, -1, 100))
margin = 1
plt.clf()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
loss = torch.clamp(x - y + margin, 0)
line, = ax.plot(x.numpy(), loss.numpy())

def update(margin=1.0):
    loss = torch.clamp(x - y + margin, 0)
    line.set_ydata(loss.numpy())
    fig.canvas.draw()

interact(update);
```
