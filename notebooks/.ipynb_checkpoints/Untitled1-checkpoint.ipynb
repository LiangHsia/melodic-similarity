```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE

%matplotlib inline

sns.set(style='white', context='paper', rc={'figure.figsize':(14,10)})
```

```python
with open('../encodings/best-pair-models_0fa9ff78_test.pkl', 'rb') as f:
    X = pickle.load(f)
tfs, ids, X = zip(*[(elt['tunefamily'], elt['id'], elt['embedding']) for elt in X])
X = np.array(X)
```

```python
import json
import gzip

id2type = {}
for line in gzip.open('../data/mtcfsinst_sequences.jsonl.gz'):
    line = json.loads(line)
    id2type[line['id']] = line['type']
```

```python
mask = np.array([id2type[id] == 'vocal' for id in ids])
```

```python
import umap
reducer = umap.UMAP(metric="cosine")
embedding = reducer.fit_transform(X)
```

```python
colors = "#2ecc71", "#1abc9c", "#f1c40f", "#e67e22", "#e74c3c", "#3498db", "#8e44ad"
labels = random.sample([label for label, count in collections.Counter(tfs).items() if count > 5], len(colors))
```

```python
import random
import collections
params = {'legend.fontsize': 15,
          'legend.handlelength': 0}
plt.rcParams.update(params)

fig, (left, right) = plt.subplots(ncols=2, figsize=(8, 3))
left.scatter(embedding[~mask, 0], embedding[~mask, 1], color="#3498db", alpha=0.5, label='INST')
left.scatter(embedding[mask, 0], embedding[mask, 1], color="#e74c3c", alpha=0.5, label='FS')
left.set_aspect('equal', 'datalim')
left.legend(bbox_to_anchor=(0., 0.5, 1.2, .102), loc="center right", frameon=False)
left.axis('off');

background_nodes = ~np.isin(np.array(tfs), labels)
right.scatter(embedding[background_nodes, 0], embedding[background_nodes, 1], color="#95a5a6", alpha=0.2)
for label, color in zip(labels, colors):
    nodes = np.array(tfs) == label
    right.scatter(embedding[nodes, 0], embedding[nodes, 1], color=color, alpha=0.7, label=label)
right.set_aspect('equal', 'datalim')
right.legend(bbox_to_anchor=(0., 0.5, 1.35, .102), loc="center right", frameon=False)
right.axis('off');
plt.tight_layout(pad=-0.7);
```

```python
import sklearn.metrics as metrics

def mean_average_precision(embs, labels, metric):
    sim_matrix = 1 - metrics.pairwise_distances(embs, metric=metric)
    scores = []
    for i, sims in enumerate(sim_matrix):
        mask = np.arange(sims.shape[0]) != i # filter query
        query_y = labels[i]
        target_y = (labels[mask] == query_y).astype(int)
        if target_y.sum() > 0:
            scores.append(metrics.average_precision_score(target_y, sims[mask]))
    return scores

aps = mean_average_precision(X, np.array(tfs), 'cosine')
```

```python
def mean_average_precision_at_1(embs, labels, metric):
    sim_matrix = 1 - metrics.pairwise_distances(embs, metric=metric)
    scores = []
    for i, sims in enumerate(sim_matrix):
        mask = np.arange(sims.shape[0]) != i # filter query
        query_y = labels[i]
        target_y = (labels[mask] == query_y).astype(int)
        best = np.argsort(sims[mask])[-1]
        scores.append(target_y[best])
    return scores

scores = mean_average_precision_at_1(X, np.array(tfs), 'cosine')
```
